```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from llama_index.core.llms import ChatMessage, MessageRole
from llama_index.llms.groq import Groq
```

```python
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
llm=Groq(
    model="llama-3.3-70b-versatile",
    temperature=0.1
)
```

```python
def chat():
    chat_history=[
        ChatMessage(role=MessageRole.SYSTEM,content="You are a helpful chatbot. Be concise and accurate")
    ]
    print("Llama index Chatbot. type 'exit' to quit \n ")
    while True:
        user_input=input("You: ").strip()
        if user_input.lower()=="exit":
            break
        print(f"You: {user_input} \n")
        chat_history.append(ChatMessage(role=MessageRole.USER,content=user_input))
        resp=llm.chat(chat_history)
        response=resp.message.content
        print(f"Bot: {response}\n")
        chat_history.append(ChatMessage(role=MessageRole.ASSISTANT,content=response))
        print("-"*80)
```

```python
chat()
```
